## ■1からnまでの総和を求める
・nはキー入力した値とする

### (1) とりあえず1から10までの繰り返し処理を作成する

```python
for i in range(1,11): # 1から10まで繰り返す
  print(i, end=' ') # iの値をスペース区切りで表示
```

### (2) 合計を格納する変数を追加して足し込む

```python
sum = 0
for i in range(1,11): # 1から10まで繰り返す
  sum += i
print(sum) # 足し込んだ値を表示する
```

### (3) nの値をキー入力する

```python
n = int(input('nの値='))
sum = 0
for i in range(1,n+1): # 1からnまで繰り返す
  sum += i
print(sum) # 足し込んだ値を表示する
```

### (4) 補足：ガウスの公式(等差数列の和)

```python
n = int(input('nの値='))
print(n * (n+1) // 2) # ガウスの公式(等差数列の和)
```

# ■FizzBuzz問題
・1から100までの数字を出力する処理で、3の倍数の時はFizz、5の倍数の時はBuzz、3と5の倍数のときはFizzBuzzと出力する

## (1) まずは1から100までの値を表示する

```python
for i in range(1, 101): # 1から100まで繰り返す
  print(i, end=' ') # 数値を表示
```

## (2) 3の倍数の時はFizzを表示する

```python
for i in range(1, 101): # 1から100まで繰り返す
  if i % 3 == 0: # 3の倍数の時はFizzを表示
    print('Fizz', end=' ')
  else:
    print(i, end=' ') # 上記以外は数値を表示
```

### (3) 5の倍数の時はBuzzを表示する

```python
for i in range(1, 101): # 1から100まで繰り返す
  if i % 3 == 0: # 3の倍数の時はFizzを表示
    print('Fizz', end=' ')
  elif i % 5 == 0: # 5の倍数の時はBuzzを表示
    print('Buzz', end=' ')
  else:
    print(i, end=' ') # 上記以外は数値を表示
```

### (4) 3と5の倍数のときはFizzBuzzと出力する
※失敗の例

```python
for i in range(1, 101): # 1から100まで繰り返す
  if i % 3 == 0: # 3の倍数の時はFizzを表示
    print('Fizz', end=' ')
  elif i % 5 == 0: # 5の倍数の時はBuzzを表示
    print('Buzz', end=' ')
  elif i % 3 == 0 and i % 5 == 0: # 3と5の倍数の時はFizzBuzzを表示
    print('FizzBuzz', end=' ')
  else:
    print(i, end=' ') # 上記以外は数値を表示
```

### (5) 3と5の倍数のときはFizzBuzzと出力する
※完成版

```python
for i in range(1, 101): # 1から100まで繰り返す
  if i % 3 == 0 and i % 5 == 0: # 3と5の倍数の時はFizzBuzzを表示
    print('FizzBuzz', end=' ')
  elif i % 3 == 0: # 3の倍数の時はFizzを表示
    print('Fizz', end=' ')
  elif i % 5 == 0: # 5の倍数の時はBuzzを表示
    print('Buzz', end=' ')
  else:
    print(i, end=' ') # 上記以外は数値を表示
```

## ■数当てゲーム
・1から10までの間で任意の数値を当てる

### (2) 入力回数を格納する変数を初期化し、開始メッセージを表示する

```python
import random # 乱数を使えるようにする
ans = random.randint(1, 10) # 1から10までの間の乱数を生成する
num = 0 # 入力回数を格納する変数
print('数当てゲーム、スタート！')
```

### (3) 永久ループを記述し、最後のメッセージを表示する

```python
import random # 乱数を使えるようにする
ans = random.randint(1, 10) # 1から10までの間の乱数を生成する
num = 0 # 入力回数を格納する変数
print('数当てゲーム、スタート！')

while True: # 永久ループ
  break

print('答えは', ans, 'です。', num, '回で正解しました！')
```

### (4) 回答をキー入力する処理と入力回数のカウントを行う

```python
import random # 乱数を使えるようにする
ans = random.randint(1, 10) # 1から10までの間の乱数を生成する
num = 0 # 入力回数を格納する変数
print('数当てゲーム、スタート！')

while True: # 永久ループ
  n = int(input('1から10までの数字を入力：')) # 解答をキー入力する
  num += 1 # 入力回数をカウントする
  break

print('答えは', ans, 'です。', num, '回で正解しました！')
```

### (5) 入力値が小さい時、大きい時、正解の時の処理を追加する

```python
import random # 乱数を使えるようにする
ans = random.randint(1, 10) # 1から10までの間の乱数を生成する
num = 0 # 入力回数を格納する変数
print('数当てゲーム、スタート！')

while True: # 永久ループ
  n = int(input('1から10までの数字を入力：')) # 回答をキー入力する
  num += 1 # 入力回数をカウントする
  if n < ans: # 乱数の値より入力値が小さいケース
    print('もっと大きいよ')
  elif n > ans: # 乱数の値より入力値が大きいケース
    print('もっと小さいよ')
  else: # 乱数の値と入力値が等しいケース
    print('正解です！')
    break

print('答えは', ans, 'です。', num, '回で正解しました！')
```
