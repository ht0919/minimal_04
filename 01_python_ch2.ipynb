## ■計算式

```python
# 算術演算子
a = 1 + 2   # 加算
b = 3 - 1   # 減算
c = 2 * 3   # 乗算
d = 4 / 2   # 除算(結果を実数で返す)
e = 10 // 3 # 除算(結果を整数で返す)
f = 10 % 3  # 剰余(あまり)
g = 10 ** 3 # べき乗

print(a, b, c, d, e, f, g)
```

```python
# 累算代入演算子
num  = 5 # 初期値として5を代入
num += 2 # 2を加算 -> 7
num -= 2 # 2を減算 -> 5
num *= 2 # 2を乗算 -> 10
num /= 2 # 2を除算 -> 5.0

print(num)
```

## ■分岐

○if文の種類
- if文、if-else文、if-elif-else文

○比較演算子の種類
- a == b → a と b は等しい
- a != b → a と b は等しくない
- a <  b → a は b より小さい
- a >  b → a は b より大きい
- a >= b → a は b 以上
- a <= b → a は b 以下


```python
a = 1
b = 1

if a == b:
  print('a と b は等しい')

if a == b:
  print('a と b は等しい')
else:
  print('a と b は等しくない')

if a > b:
  print('a は b より大きい')
elif a < b:
  print('a は b より小さい')
else:
  print('a と b は等しい')
```

## ■繰り返し

```python
# for文
for i in range(10): # 0から9まで繰り返し
  print(i, end=' ')
print() # 改行

for i in range(1,11): # 1から10まで繰り返し
  print(i, end=' ')
print() # 改行

for i in range(1, 11, 2): # 1から10まで2ステップで繰り返し
  print(i, end=' ')
print() # 改行
```

```python
# while文
i = 0
while i < 10: # 0から9まで繰り返し
  print(i, end=' ')
  i += 1
print() # 改行

i = 1
while i <11: # 1から10まで繰り返し
  print(i, end=' ')
  i += 1
print() # 改行

i = 1
while i < 11: # 1から10まで2ステップで繰り返し
  print(i, end=' ')
  i += 2
print() # 改行
```

```python
# break文
for i in range(10): # 0から9まで繰り返し
  print(i, end=' ')
  if i == 5: # iが5と等しかったら繰り返しを中断する
    break
print() # 改行

i = 0
while i < 10: # 0から9まで繰り返し
  print(i, end=' ')
  if i == 5: # iが5と等しかったら繰り返しを中断する
    break
  i += 1
print() # 改行
```
